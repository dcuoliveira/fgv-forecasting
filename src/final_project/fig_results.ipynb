```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from prettytable import PrettyTable
import numpy as np
```

```python
predict_df = pd.read_csv(os.path.join(os.path.dirname(os.getcwd()), 'data', 'df_predict_es_l1_l2_augmented.csv'))
predict_df.head()
```

```python
return_df = pd.read_csv(os.path.join(os.path.dirname(os.getcwd()), 'data', 'df_return_es_l1_l2_augmented.csv'))
return_df.head()
```

```python
curme_df = pd.read_csv(os.path.join(os.path.dirname(os.getcwd()), 'data', 'df_curme.csv'))
curme_df.head()
```

```python
return_df.plot(figsize=(15, 10), legend=False)
```

```python
mean_lasso_return = pd.concat([return_df[['date']], pd.DataFrame(return_df.mean(axis=1))], axis=1)
mean_lasso_return.set_index('date', inplace=True)
buy_n_hold_return = curme_df[['date', 'pct_change']]
buy_n_hold_return.set_index('date', inplace=True)
df1 = pd.merge(mean_lasso_return, buy_n_hold_return, right_index=True, left_index=True)
df1.columns = ['lasso_ret', 'buy_n_hold_ret']
df1['comb'] = df1.mean(axis=1)
df1.plot(figsize=(15, 10))
```

```python
df1 = pd.DataFrame(df1.mean() / df1.std())
df1.columns = ['Sharpe Ratio']
df1.to_csv(os.path.join(os.path.dirname(os.getcwd()), 'data', 'df_sharpe_ratio.csv'))
```

```python
curme_df.set_index('date').plot(figsize=(15, 10), legend=False)
```

```python
df2 = curme_df.copy()
df2.set_index('date', inplace=True)
df2 = pd.DataFrame(df2.pct_change(1).replace(np.inf, 0).replace(-np.inf, 0).median())
df2.columns = ['Median Return']
df2.to_csv(os.path.join(os.path.dirname(os.getcwd()), 'data', 'df_median.csv'))
```
